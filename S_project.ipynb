```python
import pandas as pd

# Check if 'Order_YearMonth' column is present in the DataFrame
if 'Order_YearMonth' in df.columns:
    # Group by 'Order_YearMonth' and sum 'Profit_Margin'
    result = df.groupby('Order_YearMonth')['Profit_Margin'].sum()
    
    # Convert the result to a DataFrame
    result_df = pd.DataFrame(result)
    
    # Sort the DataFrame by 'Profit_Margin'
    result_df = result_df.sort_values(by='Profit_Margin', ascending=False)
    
    print(result_df)
else:
    print("The column 'Order_YearMonth' is missing in the DataFrame.")
```
